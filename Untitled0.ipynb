```python
# modelagem e matriz
import numpy as np
import pandas as pd
```

```python
# analise grafica

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
base_Dados = pd.read_csv('PS4_GamesSales.csv', encoding='latin-1')

base_Dados.head()
```

```python
base_Dados.isnull().sum()
```

```python
plt.figure( figsize=(14,5 ))
plt.title('Verificando campos nulos')
sns.heatmap(base_Dados.isnull(), cbar=False)
```

```python
# tirando valores nulos

base_Dados.dropna(inplace=True)
```

```python
# estatisticas
base_Dados.describe()
```

```python

# Tamanho da imagem
plt.figure(figsize=(10,5))

# Titulo
plt.title ('quantidade de vendas globais (mi)', loc='left', fontsize=14)

# grafico
sns.barplot(data=base_Dados, x='Year', y='Global', ci=None, color='#689f38', estimator=sum  )

# label
plt.ylabel('quantidade vendas (mi)')
```

```python
# retirando os anos

base_Dados = base_Dados.loc[(base_Dados['Year'] !=2019 ) & (base_Dados['Year'] !=2020 ) ]


#verificar

base_Dados.head()
```

```python



# tamanho
plt.figure(figsize=(12,5))

# estilo

plt.style.use('ggplot')


# TITULO
plt.title('distribuição das vendas globais', loc='left')


# plot

sns.kdeplot(base_Dados['Global'], shade=True, bw=1, color='#689f38', linewidth=2.5);
```

```python
base_Dados.groupby(by=['Year']).sum()
```

```python

# tamanho
plt.figure(figsize=(10,4))

# titulo
plt.title('análise da distribuição Global (mi) ')

#plot
sns.boxplot(data=base_Dados, x='Year', y='Global')
```
